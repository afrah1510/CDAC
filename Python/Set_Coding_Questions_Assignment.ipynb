```python
# 1. Imagine you have a list of numbers. Write a Python code to create a set that contains only the unique numbers in the list.

num_list = [10, 20, 30, 10, 40, 50, 20, 60]
num_set = set(num_list)
print(num_set)
```

```python
# 4. You have a set of words. Write a Python code to create a new set that contains only the words that start with the letter "a".

words = {"apple", "banana", "avocado", "grape", "apricot", "melon"}
word_set = set()

for i in words:
    if i.startswith('a'):
      word_set.add(i)  

print(word_set)
```

```python
# 5. Imagine you have a set of customer IDs. Write a Python code to create a new set that contains only the IDs of customers who have made more than 5 purchases.

purchase = {"C101": 3, "C102": 7, "C103": 2, "C104": 9, "C105": 5}
cust = set()

for i, j in purchase.items():
    if j > 5:
        cust.add(i)

print(cust)
```

```python
# 8. You have a set of email addresses. Write a Python code to create a new set that contains only the email addresses that end with ".edu".

emails = {
    "student1@university.edu",
    "user2@gmail.com",
    "professor@college.edu",
    "admin@company.com"
}
email_set = set()

for i in emails:
    if i.endswith('.edu'):
        email_set.add(i)

print(email_set)
```

```python
# 13. You have a set of words. Write a Python code to create a new set that contains only the words that end with the letter "s".

word_set = {"apples", "banana", "grapes", "melon", "oranges", "kiwi"}
new_set = set()

for i in word_set:
    if i.endswith('s'):
        new_set.add(i)

print(new_set)
```

```python
# 14. Imagine you have a set of customer IDs. Write a Python code to create a new set that contains only the IDs of customers who have made at least one purchase in the last week.

all_customers = {
    ("C101", "2025-10-15"),
    ("C102", "2025-10-01"),
    ("C103", "2025-10-14"),
    ("C104", "2025-09-30"),
    ("C105", "2025-10-16")
}
recent_customer = set()

for i, j in all_customers:
    if "2025-10-09" <= j <= "2025-10-16":
        recent_customer.add(i)

print(recent_customer)
```

```python
# 15. Suppose you have a list of strings representing website URLs. Write a Python code to create a set that contains only the unique domain names of the websites.

urls = [
    "https://www.google.com",
    "https://mail.yahoo.com",
    "https://www.mit.edu",
    "https://shop.amazon.in",
    "https://www.microsoft.com"
]
domain = set()

for i in urls:
    domain.add(i.split('//')[1])

print(domain)
```

```python
# 16. You have a set of book titles. Write a Python code to find the length of the longest book title in the set.

books = {
    "Python Programming Essentials",
    "Data Science with R",
    "AI for Beginners",
    "Deep Learning Fundamentals",
    "Web Development Basics"
}
longest_word = ""
longest_len = 0

for i in books:
    if len(i) > longest_len:
        longest_word = i
        longest_len = len(i)

print(f"Longest word '{longest_word}' with length {longest_len}")
```

```python
# 17. Imagine you have two sets of words. Write a Python code to create a new set that contains only the words that are present in both sets, and also contain the letter "a".

set_x = {"data", "model", "train", "accuracy", "loss"}
set_y = {"model", "accuracy", "predict", "evaluate"}
set_z = set()

for i in set_x:
    if i in set_y and 'a' in i:
        set_z.add(i)

print(set_z)
```

```python
# 18. You have a set of email addresses. Write a Python code to create a new set that contains only the email addresses that end with ".com" and contain the word "sales" in the username.

email_set = {
    "sales_team@company.com",
    "info@business.org",
    "support@service.com",
    "sales@retail.com",
    "admin@portal.net"
}
new_email = set()

for i in email_set:
    if i.endswith('.com') and 'sales' in i:
        new_email.add(i)

print(new_email)
```

```python
# 19. Suppose you have a set of integers. Write a Python code to create a new set that contains only the integers that are greater than 50 and less than 100.

num = {10, 45, 55, 60, 75, 90, 105, 120}
new_num = set()

for i in num:
    if i > 50 and i < 100:
        new_num.add(i)

print(new_num)
```
