```python
# 1. Car Class: Basic Information Display

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Car: {self.year} {self.make} {self.model}")

c = Car("Tata", "Nexon", 2023)
c.display_info()
```

```python
# 2. Employee Bonus Calculation

class Employee:
    def __init__(self, name, age, salary):
        self.__name = name
        self.__age = age
        self.__salary = salary

    def calculate_bonus(self):
        return 0.1*self.__salary

e = Employee("Priya", 28, 60000)
print(f"Bonus: {e.calculate_bonus()}")
```

```python
# 3. BankAccount Transaction Control

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def withdraw(self, withdraw_amt):
        self.withdraw_amt = withdraw_amt
        if self.withdraw_amt > self.__balance:
            print("Withdrawal failed. Insufficient funds")
        elif self.withdraw_amt < 0:
            print("Withdraw amount cannot be negative")
        else:
            if self.__balance < 3000:
                self.__balance -= 500
                print("Penalty deducted Rs. 500")
                print("New Balance:", self.__balance)
            self.__balance -= self.withdraw_amt
            print("New Balance:", self.__balance)

    def deposit(self, deposit_amt):
        self.deposit_amt = deposit_amt
        if self.deposit_amt < 0:
            print("Deposit amount cannot be negative")
        else:
            self.__balance += self.deposit_amt
        print(f"New amount = {self.__balance}")
acc = BankAccount(1000)
acc.withdraw(1200)
acc.deposit(500)
```

```python
# Create account only if user enters

while True:
    amount = float(input("Enter amount 3000+ to open account (0 to quit): "))
    if amount >= 3000:
        acc = BankAccount(amount)
        print("Account created")
        break
    elif amount == 0:
        break
    else:
        print("Minimum 3000 needed to open account, you entered {amount}")

acc.withdraw(1200)
acc.deposit(500)
```

```python
# 4. Student Grade Analysis

class Student:
    def __init__(self, name, *grades):
        self.__name = name
        self.__grades = grades

    def __str__(self):
        return f"{self.__name} and grades are {self.__grades}"

    def calculate_average(self):
        return sum(self.__grades) / len(self.__grades)

    def find_highest_grade(self):
        return max(self.__grades)

    def find_lowest_grade(self):
        return min(self.__grades)

s = Student("Amar", 88, 92, 75, 95)
print(s)
print(f"Average: {s.calculate_average()}")
print(f"Highest grade: {s.find_highest_grade()}")
print(f"Lowest grade: {s.find_lowest_grade()}")
```
