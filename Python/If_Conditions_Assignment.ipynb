```python
# 1. Write a program that takes a number as input and prints out whether it is an even or odd number.

num = int(input("Enter a number: "))
if num%2 == 0:
    print(num, "is even number")
else:
    print(num, "is odd number")
```

```python
# 2. Write a program that takes a year as input and prints out whether it is a leap year or not.

year = int(input("Enter a year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap year")
        else:
            print("Not a leap year")
    else:
        print("Leap year")
else:
    print("Not a leap year")
```

```python
# 3. Write a program that takes a string as input and checks whether it is a palindrome or not.

str1 = input("Enter a string: ")
rev_str = ""

for char in str1:
    rev_str = char + rev_str

if str1 == rev_str:
    print(str1, "is a palindrome")
else:
    print(str1, "is not a palindrome")
```

```python
# 4. Write a program that takes three numbers as input and prints out the largest one.

num1, num2, num3 = map(int, input("Enter three numbers separated by space(' '): ").split())
if num1 > num2 and num1 > num3:
    print(num1, "is the largest")
elif num2 > num1 and num2 > num3:
    print(num2, "is the largest")
else:
    print(num3, "is the largest")
```

```python
# 5. Write a program that takes a sentence as input and counts the number of words in it.

sentence = input("Enter a sentence: ").split()
count = 0
for i in sentence:
    if i == " ":
        continue
    count += 1

print("Number of words in the sentence:", count)
```

```python
# 6. Write a program that takes a list of numbers as input and prints out the sum of all the numbers.

num1, num2, num3 = map(int, input("Enter three numbers separated by space(' '): ").split())
num_list = [num1, num2, num3]
sum = 0
for i in num_list:
    sum += i

print("Sum of all numbers in list:", sum)
```

```python
# 7. Write a program that takes a list of strings as input and prints out the longest string.

str_list = list(map(str, input("Enter three strings separated by space(' '): ").split()))

for i in str_list:
    if len(str_list[0]) > len(str_list[1]) and len(str_list[0]) > len(str_list[2]):
        print("Longest string in list:", str_list[0])
        break
    elif len(str_list[1]) > len(str_list[0]) and len(str_list[1]) > len(str_list[2]):
        print("Longest string in list:", str_list[1])
        break
    else:
        print("Longest string in list:", str_list[2])
        break
```

```python
# 8. Write a program that takes two lists of numbers as input and prints out the common elements between the two lists.

num_list1 = list(map(int, input("Enter numbers for list 1 separated by space: ").split()))
num_list2 = list(map(int, input("Enter numbers for list 2 separated by space: ").split()))
common_element = []
for i in num_list1:
    for j in num_list2:
        if i == j:
            common_element.append(i)

print("Common elements between two lists:", common_element)
```

```python
# 9. Write a program that takes a list of numbers as input and prints out the numbers in ascending order.

num_list = list(map(int, input("Enter numbers for list separated by space: ").split()))
for i in range(len(num_list)):
    min_index = i
    for j in range(i+1, len(num_list)):
        if num_list[j] < num_list[min_index]:
            min_index = j
    num_list[i], num_list[min_index] = num_list[min_index], num_list[i]

print("List in ascending order:", num_list)
```

```python
# 10. Write a program that takes a list of strings as input and removes all duplicates from the list.

str_list = list(map(str, input("Enter three strings separated by space(' '): ").split()))

for i in str_list:
    if str_list.count(i) > 1:
        str_list.remove(i)

print("List without duplicates:", str_list)
```

```python
# 11. Write a program that takes a string as input and prints out the frequency of each character in the string.

str1 = input("Enter a string: ")
freq = {}

for char in str1:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

for i, j in freq.items():
    print(i, "-->", j)
```

```python
# 13. Write a program that takes a list of numbers as input and prints out the median value.

num_list = list(map(int, input("Enter numbers for list separated by space: ").split()))
sort_list = sorted(num_list)
n = len(num_list)

if n%2 == 0:
    mid1 = sort_list[n//2 -1]
    mid2 = sort_list[n//2]
    median = (mid1 + mid2) / 2
else:
    median = sort_list[n//2]

print("Median:", median)
```

```python
# 20. Write a program that takes a list of numbers as input and checks whether it is a Fibonacci sequence or not.

num_list = list(map(int, input("Enter numbers for list separated by space: ").split()))
if len(num_list) <= 2:
    is_fibonacci = False
else:
    is_fibonacci = True
    for i in range(2, len(num_list)):
        if num_list[i] != num_list[i-1] + num_list[i-2]:
            is_fibonacci = False
            break

if is_fibonacci:
    print("Fibonacci sequence")
else:
    print("Not a Fibonacci sequence")
```

```python
# 25. Write a program that takes a string as input and prints out the most common word in the string.

str_list = list(map(str, input("Enter strings for list separated by space: ").lower().split()))
freq = {}

for i in str_list:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
for i, j in freq.items():
    if j == max(freq.values()):
        print("Most common word:", i, ", with count:", j)
```

```python
# 27. Write a program that takes a string as input and removes all the vowels from the string.

str_list = list(map(str, input("Enter strings for list separated by space: ").lower().split()))
vowel = "aeiou"
result = []

for i in str_list:
    new = ""
    for char in i:
        if char not in vowel:
            new += char
    result.append(new)

print(" ".join(result))
```

```python
# 28. Write a program that takes a list of numbers as input and checks whether it is a prime number sequence or not.

num_list = list(map(int, input("Enter numbers for list separated by space: ").split()))
is_prime = True

for i in num_list:
    if i < 2:
        is_prime = False 
        break
    for j in range(2, i):
        if i%j == 0:
            is_prime = False
            break

if is_prime == True:
    print("Prime number sequence")
else:
    print("Not a prime number sequence")
```

```python
# 32. Write a program that takes a string as input and prints out the number of uppercase letters and lowercase letters in the string.

str1 = input("Enter a string: ")
u_count = 0
l_count = 0

for i in str1:
    if i.isupper():
        u_count += 1
    if i.islower():
        l_count += 1

print("Number of uppercase letters:", u_count)
print("Number of lowercase letters:", l_count)
```

```python
# 37. Write a program that takes a string as input and prints out the number of times each word appears in the string.

text = '''Knowing your own darkness is the best method for dealing with the darknesses of other people. 
The privilege of a lifetime is to become who you truly are. 
People will do anything, no matter how absurd, to avoid facing their own souls. 
The most terrifying thing is to accept oneself completely. 
How can I be substantial if I do not cast a shadow? I must have a dark side also If I am to be whole.
'''

text = text.replace(",", "")
text = text.replace(".", "")
text = text.replace("?", "")

count_freq = {}

for i in text.lower().split():
    if i not in count_freq:
        count_freq[i] = 1
    else:
        count_freq[i] += 1

print("Number of times each word appears are - ")
for i, j in count_freq.items():
    print(i, "--->", j)
```

```python
# 39. Write a program that takes a list of numbers as input and prints out a new list where each element is the product of all the elements in the original list except for the current element.

num_list = list(map(int, input("Enter numbers for list separated by space: ").split()))
product_num = []
pro = 1

for i in num_list:
    pro *= i

for i in num_list:
    num = pro/i
    product_num.append(num)

print("Product List:", product_num)
```
