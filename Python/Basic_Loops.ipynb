```python
# 1. Count Vowels

my_str = input("Enter a string: ")
count = 0
for c in 'aeiou':
  print(c, my_str.lower().count(c))
  count += my_str.count(c)

print("Total number of vowels: ", count)
```

```python
# 2. Reverse String

my_str = input("Enter a string: ")
rev = ""
for i in range(len(my_str)-1, -1, -1):
  rev += my_str[i]

print(rev)
```

```python
# 3. Check Palindrome

my_str = input("Enter a string: ").lower()
if my_str == my_str[::-1]:
  print(my_str, "is a palindrome")
else:
  print(my_str, "is not a palindrome")
```

```python
# 4. Word Counter

my_str = input("Enter a line: ").split()
print("Number of words are: ", len(my_str))
```

```python
# 5. Remove Spaces

my_str = input("Enter a string: ")
new = ""
for i in my_str:
  if i == " ":
    continue
  new += i

print(new)
```

```python
# 6. Title Case

my_str = input("Enter a string (in lowercase): ").split()
new = ""

for i in my_str:
  new += i[0].upper() + i[1:].lower() + " "

print(new)
```

```python
# 7. Toggle Case

my_str = input("Enter a string: ")
new = ""

for i in my_str:
  if i.isupper() == True:
    new += i.lower()
  if i.islower() == True:
    new += i.upper()
  if i == " ":
    new += " "

print(new)
```

```python
# 8. Caesar Cipher (Simple)

my_str = input("Enter a string: ")
key = int(input("Enter key: "))
cipher = ""

for i in my_str.lower():
  if i.isalpha():
    cipher += chr((ord(i)-ord('a')+key)%26 + ord('a'))
  else:
    cipher += i

print(cipher)
```

```python
# 9. Remove Duplicates (Preserve Order)

my_str = input("Enter a string: ").lower()
new = ""

for i in my_str:
  if i not in new:
    new += i

print(new)
```

```python
# 10. Count Character Frequency

my_str = input("Enter a string: ")
char_freq = {}

for i in my_str.lower():
  if i in char_freq:
    char_freq[i] += 1
  else:
    char_freq[i] = 1

print(char_freq)
```

```python
# 11. Find First Substring Index

my_str = input("Enter a string: ")
sub = input("Enter substring: ")
match = False

for i in range(len(my_str)):
  if my_str[i:len(sub)+i] == sub:
    match = True
    print("Substring found at", i)
    break

if match == False:
  print("-1")
```

```python
# 12. Extract Middle Character(s)

my_str = input("Enter a string: ")
char = ""

if len(my_str) % 2 == 0:
  char = my_str[len(my_str)//2 - 1] + my_str[len(my_str)//2]
else:
  char = my_str[len(my_str)//2]

print(char)
```

```python
# 13. Remove Nth Index

my_str = input("Enter a string: ")
index = int(input("Enter index number to be removed (0-based): "))
new = my_str[:index]+my_str[index+1:]
print("After removal of index number", index, ":", new)
```

```python
# 14. Check Anagrams

my_str1 = list(input("Enter string 1: ").lower())
my_str2 = list(input("Enter string 2: ").lower())
freq1 = {}
freq2 = {}

for i in my_str1:
  if i in freq1:
    freq1[i] += 1
  else:
    freq1[i] = 1

for j in my_str2:
  if j in freq2:
    freq2[j] += 1
  else:
    freq2[j] = 1

if freq1 == freq2:
  print("Strings are anagrams")
else:
  print("Strings are not anagrams")
```

```python
# 15. Longest Word

my_str = list(input("Enter a string: ").split())
longest = ""

for i in range(len(my_str)):
  if len(my_str[i]) > len(longest):
    longest = my_str[i]

print("Longest word:", longest)
print("Length:", len(longest))
```

```python
# 16. Check for only digits

my_num = input("Enter a number: ")
flag = True

for i in my_num:
  if not (48 <= ord(i) and ord(i) <= 57):
    flag = False
    break

if flag == True:
  print("True")
else:
  print("False")
```

```python
# 17. String Compression

my_str = input("Enter a string: ")
op = ""
count = 1

for i in range(1, len(my_str)):
  if my_str[i] == my_str[i-1]:
    count += 1
  else:
    op += my_str[i-1]+str(count)
    count = 1
op += my_str[i]+str(count)

print(op)
```

```python
# 18. Check Prefix/Suffix

my_str = input("Enter a string: ")
prefix = input("Enter a prefix to be checked: ")
suffix = input("Enter a suffix to be checked: ")

if prefix == "":
  print("Prefix: False")
elif len(prefix) > len(my_str):
  print("Prefix bigger than string")
elif prefix == my_str[:len(prefix)]:
  print("Prefix: True")
else:
  print("Prefix: False")

if suffix == "":
  print("Suffix: False")
elif len(suffix) > len(my_str):
  print("Suffix bigger than string")
elif suffix == my_str[len(my_str)-len(suffix):]:
  print("Suffix: True")
else:
  print("Suffix: False")
```

```python
# 19. Interleave Strings

my_str1 = input("Enter a string: ")
my_str2 = input("Enter a string: ")
new = ""

if len(my_str1) < len(my_str2):
  for i in range(len(my_str1)):
    new += my_str1[i]
    for j in range(i, i+1):
      new += my_str2[j]
  new += my_str2[len(my_str1):]
else:
  for i in range(len(my_str2)):
    new += my_str1[i]
    for j in range(i, i+1):
      new += my_str2[j]
  new += my_str1[len(my_str2):]

print(new)
```

```python
# 20. Simple String Calculator

my_str = input("Enter your expression (each element separated by space): ")
num1, op, num2 = my_str.split()

if op == "+":
  result = int(num1) + int(num2)
elif op == "-":
  result = int(num1) - int(num2)
elif op == "*":
  result = int(num1) * int(num2)
elif op == "/":
  result = int(num1) / int(num2)
else:
  print("Invalid operator")

print("Result:", result)
```
