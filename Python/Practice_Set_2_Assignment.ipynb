```python
# 3. Write a Python program to create a Vehicle class with max_speed and mileage instance Attributes.

class Vehicle:
    def __init__(self, max_speed, mileage):
        self.max_speed = max_speed
        self.mileage = mileage
    
    def display_info(self):
        return f"Max Speed: {self.max_speed} kmph, Mileage: {self.mileage} kmpl"

v = Vehicle(5, 100)
v.display_info()
```

```python
# 8. Define a function overlapping() that takes two lists and returns True if they have at least one member in common, False otherwise.

def overlapping(num_list1, num_list2):
    common = []
    for i in num_list1:
        if i in num_list2:
            common.append(i)
    
    if len(common) >= 1:
        return True
    else:
        return False

num_list1 = [93, 64, 80, 37, 20]
num_list2 = [68, 4, 93, 67, 46]
overlapping(num_list1, num_list2)
```

```python
# 9. Write a function find_longest_word() that takes a list of words and returns the length of the longest word and that word itself.

def find_longest_word(str_list):
    longest = ""
    for i in str_list:
        if len(i) > len(longest):
            longest = i
    return f"Longest word: '{longest}' with length {len(longest)}"

str_list = ["Aileen", "Dante", "Fire", "Pinwheel", "Aju Nice", "Madam", "Kayak", "D.I.D"]
find_longest_word(str_list)
```

```python
# 10. Write a function filter_long_words() that takes a list of words and an integer len and returns the list of words that are longer than len.

def filter_long_words(str_list, int_len):
    new_list = []
    for i in str_list:
        if len(i) > int_len:
            new_list.append(i)
    
    return new_list

str_list = ["Aileen", "Dante", "Fire", "Pinwheel", "Aju Nice", "Madam", "Kayak", "D.I.D"]
filter_long_words(str_list, 5)
```

```python
# 16. Use inheritance:
'''
Write a class Course with name, fees.
Provide following functionalities:
A. initializer
B. Accept data
C. Print Data
Write a Class Computer with subjectList.
Provide following functionalities:
A. initializer
B. Accept data
C. Print Data
Write a Class Electronics with subjectList.
Provide following functionalities:
A. initializer
B. Accept data
C. Print Data
Write menu driven program to test above functionalities.
Output: ‘1st year Comp’ 90000 [‘sub1’,’sub2’....]
'''

class Course:
    def __init__(self):
        self.name = ""
        self.fees = 0
    def accept_data(self, name, fees):
        self.name = name
        self.fees = fees
    def print_data(self):
        print(f"Name: {self.name}, Fees: {self.fees}")

class Computer(Course):
    def __init__(self):
        Course.__init__(self)
        self.subject = []
    def accept_data(self, name, fees, subject):
        self.name = name
        self.fees = fees
        self.subject = subject
    def print_data(self):
        print(f"Name: {self.name}, Fees: {self.fees}, Subject list: {self.subject}")

class Electronics(Course):
    def __init__(self):
        Course.__init__(self)
        self.subject = []
    def accept_data(self, name, fees, subject):
        self.name = name
        self.fees = fees
        self.subject = subject
    def print_data(self):
        print(f"Name: {self.name}, Fees: {self.fees}, Subject list: {self.subject}")
        
c = Course()
c.accept_data("Computer", 90000)
c.print_data()

co = Computer()
co.accept_data('AI&DS', 160000, ['Python', 'Data Mining & Business Intelligence'])
co.print_data()

e = Electronics()
e.accept_data("EXTC", 105000, ["Microprocessor", "Embedded System Design"])
e.print_data()
```

```python
# 17. Write a program that contains a function that has one parameter, n, representing an integer greater than 0. The function should return n! (n factorial). Then write a main function that calls this function with the values 1 through 20, one at a time, printing the returned results.

def fact(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return n * fact(n-1)

for i in range(1, 21):
    print(f"Factorial of {i}: {fact(i)}")
```

```python
# 18. Write a version of a palindrome recognizer that also accepts phrase palindromes such as "Go hang a salami I'm a lasagna hog.", "Was it a rat I saw?", "Step on no pets", "Sit on a potato pan, Otis", "Lisa Bonet ate no basil", "Satan, oscillate my metallic sonatas", "I roamed under it as a tired nude Maori", "Rise to vote sir", or the exclamation "Dammit, I'm mad!". Note that punctuation, capitalization, and spacing are usually ignored.

str_list = ["Go hang a salami I'm a lasagna hog.", "Was it a rat I saw?", "Step on no pets", 
            "Sit on a potato pan, Otis", "Lisa Bonet ate no basil", "Satan, oscillate my metallic sonatas", 
            "I roamed under it as a tired nude Maori", "Rise to vote sir", "Dammit, I'm mad!",
           "This world is really cruel", "Let's love even our hidden sorrows", "Be my spring"]

palindrome_str = []

for i in str_list:
    clean = i.replace(" ", "").replace(".", "").replace(",", "").replace("?", "").replace("'", "").replace("!", "").lower()

    mid = len(clean)//2
    if len(clean)%2 == 0:
        if sorted(clean[0:mid]) == sorted(clean[mid::]):
            palindrome_str.append(i)
    else:
        if sorted(clean[0:mid]) == sorted(clean[mid+1::]):
            palindrome_str.append(i)

print("Strings which are palindromes:", palindrome_str)
```

```python
# 20. Create a class 'Student' with roll no, studentName, course, dictionary of marks (subjectName -> marks [5]).
'''
Provide following functionalities:
A. initializer
B. override __str__ method
C. accept student data
D. Print student data for given id
E. Print Student who has failed in any subject
Write menu driven program to test above functionalities (accept records of 5 students and store those in list)
'''

class Student:
    def __init__(self):
        self.roll = 0
        self.name = ""
        self.course = ""
        self.marks = {}
    def __str__(self):
        return f"Roll No: {self.roll}, Name: {self.name}, Course: {self.course}, Marks: {self.marks}"
    def accept_data(self, roll, name, course, marks):
        self.roll = roll
        self.name = name
        self.course = course
        self.marks = marks
    def print_student(self, id):
        if self.roll == id:
            print(self)
    def fail_student(self):
        for subject, mark in self.marks.items():
            if mark < 40:
                return True
        return False

s1 = Student()
s2 = Student()
students = [s1, s2]

s1.accept_data(201, "Ishaan", "Computer Engineering", {"Data Structures": 80, "Operating Systems": 76, "Networks": 88, "DBMS": 91})
s2.accept_data(203, "Raghav", "Cybersecurity", {"Cryptography": 72, "Network Security": 39, "Ethical Hacking": 80, "Forensics": 70})

for s in students:
    s.print_student(201)

print("Failed Students:")
for s in students:
    if s.fail_student() == True:
        print(s)
```

```python
# 22. Write a Class Painting with the method calculatePaintingCost.
'''
Write a Class FlatPainting with noOfRooms which is inherited from Class Painting.
Override calculatePaintingCost (Assume per room painting cost is 10000).
Write a Class BuildingPainting with noOfFlats which is inherited from Class Painting.
Override calculatePaintingCost (Assume per room painting cost is 25000).
Write main() method to accept painting assignment with option 1. Flat 2. Building and then calculate the cost according to the type.
'''

class Painting:
    def calculatePaintingCost(self):
        pass  # Base method to be overridden


class FlatPainting(Painting):
    def __init__(self, noOfRooms):
        self.noOfRooms = noOfRooms

    def calculatePaintingCost(self):
        return self.noOfRooms * 10000


class BuildingPainting(Painting):
    def __init__(self, noOfFlats):
        self.noOfFlats = noOfFlats

    def calculatePaintingCost(self):
        return self.noOfFlats * 25000


def main():
    while True:
        print("1. Flat")
        print("2. Building")
        print("3. Exit")
        choice = int(input("Enter your choice: "))
    
        if choice == 1:
            rooms = int(input("Enter number of rooms: "))
            flat = FlatPainting(rooms)
            print("Total Painting Cost:", flat.calculatePaintingCost())

        elif choice == 2:
            flats = int(input("Enter number of flats: "))
            building = BuildingPainting(flats)
            print("Total Painting Cost:", building.calculatePaintingCost())

        elif choice == 3:
            print("Thank for choosing our service")
            break
        
        else:
            print("Invalid choice")
            break

main()
```
