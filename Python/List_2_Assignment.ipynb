```python
# 31. Write a Python program that takes a list of integers as input and returns a new list containing only the numbers that are perfect cubes.

num_list = [8, 64, 100, 37, 127, -343]
new_list = []

for i in num_list:
    if i < 0:
        n = round(abs(i)**(1/3))
        if n**3 == abs(i) and (n*-1)**3 == i:
            new_list.append(i)
    else:
        n = round(i**(1/3))
        if n*n*n == i:
            new_list.append(i)

print("Perfect cubes:", new_list)
```
