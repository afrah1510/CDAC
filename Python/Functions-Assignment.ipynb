```python
# 5. Maximum of three numbers
def maximum(no1, no2):
    return no1 if no1>no2 else no2

def max_of_three(a, b, c):
    return a if a>maximum(b,c) else maximum(b,c)

a = int(input("Enter a number: "))
b = int(input("Enter a number: "))
c = int(input("Enter a number: "))

print("Max:", max_of_three(a, b, c))
```

```python
# 9. Factorial using loop

def factorial_iter(n):
    if n < 0:
       return "Error"
    elif n == 0:
        return 1

    result = 1
    for i in range(1, n+1):
        result *= i
    return result

n = int(input("Enter a number: "))
print(f"Factorial of {n}: {factorial_iter(5)}")
```

```python
# 13. Remove duplicates from a list

def remove_dups(lst):
    dup_list = []
    for i in lst:
        if i not in dup_list:
            dup_list.append(i)
    return dup_list

num = input("Enter numbers separated by space: ")
lst = list(map(int, num.split()))
remove_dups(lst)
```

```python
# 14. Second largest number in a list (distinct)

def second_largest(lst):
    st = set(lst)
    n_lst = sorted(st, reverse = True)
    return n_lst[1]

num = input("Enter numbers separated by space: ")
lst = list(map(int, num.split()))
second_largest(lst)
```

```python
# 16. Count frequency of each character in a string

def count_freq(s):
    freq = {}
    for i in s.lower():
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1
    return freq

s = input("Enter a string: ")
count_freq(s)
```

```python
# 18. Merge two sorted lists (like merge step of merge-sort)

def merge_sorted(a, b):
    a_sort = sorted(a)
    b_sort = sorted(b)
    m_list = sorted(a_sort + b_sort)
    return m_list

a = input("Enter numbers for list 1 separated by space: ")
a_lst = map(int, a.split())
b = input("Enter numbers for list 2 separated by space: ")
b_lst = map(int, b.split())
merge_sorted(a_lst, b_lst)
```

```python
# 21. Factorial using recursion

def factorial_rec(n):
    if n < 0:
        return "Error"
    elif n == 1:
        return 1
    else:
        return n*factorial_rec(n-1)

n = int(input("Enter a number: "))
factorial_rec(n)
```

```python
# 22. Nth Fibonacci number (recursive or iterative)

def fibonacci(n):
    if n == 0:
        return 0

    if n == 1:
        return 1

    else:
        return fibonacci(n-1)+fibonacci(n-2)

n = int(input("Enter the nth value to access Fibonacci number: "))
fibonacci(n)
```

```python
# 23. Sum of digits (recursively)

def sum_digits(n):
    if n == 0:
        return 0
    else:
        return n%10 + sum_digits(n//10)

n = int(input("Enter a number: "))
sum_digits(n)
```

```python
# 24. Power function using recursion (x^y)

def power(x, y):
    if y == 0:
        return 1

    if y < 0:
        return 1/power(1, abs(y))

    if y%2 == 0:
        return power(x, y/2) * power(x, y/2)
    else:
        return x * power(x, y-1)

x = int(input("Enter value for x: "))
y = int(input("Enter value for y: "))
power(x, y)
```

```python
# 25. GCD of two numbers (Euclidean algorithm, recursive)

def gcd(a, b):
    a, b = abs(a), abs(b)
    if b != 0:
        return gcd(b, a%b)
    else:
        return a

a = int(input("Enter a number: "))
b = int(input("Enter a number: "))
gcd(a, b)
```

```python
# 26. Recursively print all elements of a limit

def print_list_recursive(lst, i=0):
    while i < len(lst):
        print(lst[i])
        i += 1
        return print_list_recursive(lst, i)

num = input("Enter numbers separated by space: ")
lst = list(map(int, num.split()))
print_list_recursive(lst)
```

```python
# 28. Function that accepts another function (higher-order)

def call_fn():
    return lambda x : x*2

def apply_fn(fn, lst):
    return [fn(x) for x in lst]

fn = call_fn()
num = input("Enter numbers separated by space: ")
lst = map(int, num.split())
apply_fn(fn, lst)
```

```python
# 29. Demonstrate global vs local variables
'''
First, we declare the global variable (counter)
outside the function block
Inside inc(), 'global' is used to modify the counter variable
local_var is declared inside the function block
'''

counter = 0
print("Global counter before function call:", counter)

# Will give error as local_var cannot be accessed outside the function block
# print("Local Variable called outside the function block:", local_var)

def inc():
    global counter
    counter += 1
    local_var = 5
    return counter, local_var

result = inc()
print("Global Variable, Local Variable:", result)
print("Global counter after function call:", counter)
```
