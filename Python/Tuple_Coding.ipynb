```python
# Grading System
'''
Tuples are immutable, and hence once created they can't be tampered with, making it secure and consistent
Tuples are faster and uses less memory
'''
# students = [(name, marks1, marks2, marks3, grade)]

students = [
    ("Ishaan", 85, 90, 88, "B"),
    ("Meera", 70, 75, 72, "C"),
    ("Kabir", 60, 65, 62, "D"),
    ("Johnson", 99, 98, 97, "A")
]

print("Students sorted by grades - ")
print(sorted(students, key = lambda x : x[4]))
```

```python
# Student Name and Grade

'''
scores = [(name, marks)]
results = [(name, grade)]
'''
def student_grade(scores):
    result = []
    for name, score in scores:
        if score >= 90:
            grade = "A"
        elif score >= 80:
            grade = "B"
        elif score >= 70:
            grade = "C"
        elif score >= 60:
            grade = "D"
        else:
            grade = "F"
        result.append((name, grade))
    return result
    
scores = [("Ishaan", 92), ("Johnson", 99), ("Kabir", 59), ("Meera", 76)]
print("Student with grades -")
for i in student_grade(scores):
    print(i)
```

```python
# Most Frequent Word

# words = [(word, count)]

def most_frequent(words):
    return max(words, key = lambda x : x[1])

words = [("data", 15), ("python", 22), ("model", 18)]
most_frequent(words)
```
