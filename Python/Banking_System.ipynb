```python
# Banking System

class Account:
    __acc_no = 1000
    def __init__(self, name, gender, amount):
        self.__name = name
        self.__gender = gender
        self.__balance = amount
        Account.__acc_no += 1
        self.__account_number = Account.__acc_no
        print(f"Account created with account number: {self.__account_number}")

    def get_acc_no(self):
        return self.__account_number

    def deposit(self, deposit_amount):
        if deposit_amount < 0:
            print("Deposite amount cannot be negative")
        else:
            self.__balance += deposit_amount
            print("Deposit successful")
        print(f"New Balance: {self.__balance}")

    def withdraw(self, withdraw_amount):
        if withdraw_amount < 0:
            print("Withdraw amount cannot be negative")
        elif self.__balance - withdraw_amount < 500:
            self.__balance -= 500
            print("Penalty of Rs. 500 applied as withdrawal caused balance to be less than Rs. 500")
            self.__balance -= withdraw_amount
            if self.__balance < 0:
                self.__balance += withdraw_amount
                self.__balance += 500
                print("Withdrawal denied")
            else:
                print("Withdrawal successful")
                print(f"New Balance: {self.__balance}")
        elif withdraw_amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= withdraw_amount
            print("Withdraw successful")
            print(f"New Balance: {self.__balance}")

    def transfer(self, transfer_amount):
        if transfer_amount < 0:
            print("Transfer amount cannot be negative")
        elif self.__balance - transfer_amount < 500:
            self.__balance -= 500
            print("Penalty of Rs. 500 applied as transfer caused balance to be less than Rs. 500")
            self.__balance -= transfer_amount
            if self.__balance < 0:
                self.__balance += transfer_amount
                self.__balance += 500
                print("Transfer denied")
            else:
                print("Transfer successful")
                print(f"New Balance: {self.__balance}")
        elif transfer_amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= transfer_amount
            print("Transfer successful")
            print(f"New Balance: {self.__balance}")

    def __str__(self):
        return f"Account Number: {self.__account_number}, Name: {self.__name}, Gender: {self.__gender}, Balance: {self.__balance}"
```
