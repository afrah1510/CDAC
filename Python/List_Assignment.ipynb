```python
# 1. Write a Python program to remove duplicates from a list.

num_list = list(map(int, input("Enter numbers separated by space: ").split()))
new_list = []

for i in num_list:
    if i not in new_list:
        new_list.append(i)

print("List without duplicates:", new_list)
```

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all the even numbers in the list. 
# You can assume that the input list only contains integers.

def sum_list(num_list):
    even_sum = 0
    for i in num_list:
        if i%2 == 0:
            even_sum += i
    return even_sum

num_list = list(map(int, input("Enter numbers separated by space: ").split()))
print("Sum of all elements in the list:", sum_list(num_list))
```

```python
# 2. Write a Python function that takes two lists as input and returns a new list that contains all the elements 
# that are common between the two input lists. You can assume that the input lists only contain integers, 
# and that the output list should not contain any duplicate elements.

def common_list(num_list1, num_list2):
    common_num = []
    for i in num_list1:
        if i in num_list2:
            common_num.append(i)
    return common_num

num_list1 = list(map(int, input("Enter numbers for list 1 separated by space: ").split()))
num_list2 = list(map(int, input("Enter numbers for list 2 separated by space: ").split()))
print("Common elements between the two lists:", common_list(num_list1, num_list2))
```

```python
# 3. Write a Python program that takes a list of strings as input and returns a new list containing 
# only the strings that are longer than five characters. 
# Your program should define a function filter_long_strings() that takes the 
# input list as an argument and returns the filtered list.

def filter_long_strings(str_list):
    filter_list = []
    for i in str_list:
        if len(i) > 5:
            filter_list.append(i)
    return filter_list

str_list = list(map(str, input("Enter numbers separated by space: ").split()))
print("List with strings that are longer than five characters:", filter_long_strings(str_list))
```
