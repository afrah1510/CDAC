```python
# Tea Stall Billing System - Project Requirements and Sample Run

order_dict = {
    "cutting chai" : 15,
    "masala chai" : 25,
    "samosa" : 20,
    "vada pav" : 30,
    "biscuit packet" : 10
}

gst = 0.05
orders = {}

while True:
  print("\n")
  print("-"*80)
  print("Menu".center(80, " "))
  print("-"*80)
  print("1. Cutting Chai")
  print("2. Masala Chai")
  print("3. Samosa")
  print("4. Vada Pav")
  print("5. Biscuit Packet")
  print("-"*80)

  order_input = input("Enter item name and quantity separated by '-': ").lower()
  order_list = order_input.split("-")
  item = 0
  qty = 1

  if order_input == "":
    break
  for i in range(0, len(order_list)-1, 2):
    if order_list[i] not in ("cutting chai", "masala chai", "samosa", "vada pav", "biscuit packet"):
      print("Please check the option")
    orders[order_list[i]] = orders.get(order_list[i], 0) + int(order_list[i+1])
    if i > len(order_list) or (i+1) > len(order_list):
      break

print("\n")
print("-"*80)
print("Membership Type")
print("1. Gold Membership")
print("2. Silver Membership")
print("3. No Membership")
mem = input("Enter membership type (Gold, Silver, No): ").lower()
if mem == "gold":
  discount = 0.1
elif mem == "silver":
  discount = 0.05
elif mem == "no":
  discount = 0

total = 0
print("\n\n")
print("="*80)
print("TEA STALL BILLING SYSTEM".center(80, " "))
print("="*80)
print(f"{'ITEM'.ljust(20)}{'QTY'.ljust(20)}{'PRICE'.ljust(20)}{'TOTAL'.ljust(20)}")
print("-"*80)
for item, qty in orders.items():
  price = order_dict[item]
  item_total = price*int(qty)
  total += item_total
  print(f"{item.title().ljust(20)}{str(qty).ljust(20)}{format(price, '.2f').ljust(20)}{format(item_total, '.2f').ljust(20)}")

discount_amt = discount * total
taxable_amt = total - discount_amt
gst_amt = gst * taxable_amt
payable_amt = taxable_amt + gst_amt

print("\n")
print("-"*80)
print(f"{'SUBTOTAL'.ljust(59)} {total:.2f}")
print(f"{'MEMBERSHIP:'.ljust(59)} {mem.upper()} ({discount*100:.2f}% Off)")
print(f"{'DISCOUNT APPLIED'.ljust(59)} -{discount_amt:.2f}")
print("-"*80)
print(f"{'TAXABLE TOTAL'.ljust(59)} {taxable_amt:.2f}")
print(f"{'GST ('}{gst*100:.2f}{'%)' + ''.ljust(48)} +{gst_amt:.2f}")
print("-"*80)
print(f"{'TOTAL AMOUNT DUE'.ljust(59)} {payable_amt:.2f}")
print("="*80)
print("THANK YOU! VISIT AGAIN!")
```
