```python
# 1. Write a program that prompts the user to enter a sentence and then prints out each word in uppercase. For example, if the user enters "heallo world", the program should print out "HELLO WORLD".

sentence = input("Enter a sentence: ")
new_sentence = ""

for i in sentence:
    new_sentence += i.upper()

print(new_sentence)
```

```python
# 3. Write a function that takes a string as input and returns a list of all the words in the string that are longer than a specified length. The function should take two arguments: the string and the minimum length of a word. For example, if the input string is "The quick brown fox jumps over the lazy dog" and the minimum length is 4, the function should return ["quick", "brown", "jumps"].

def words_longer(str_list, specified_length):
    new_list = []
    for i in str_list:
        if len(i) > specified_length:
            new_list.append(i)
    return new_list

str_list = list(map(str, input("Enter a string: ").split()))
specified_length = int(input("Enter a minimum length: "))
words_longer(str_list, specified_length)
```

```python
# 4. Write a program that reads in a list of names from the user, separated by commas, and then prints out the number of names that begin with a vowel. For example, if the user enters "Anna, Bob, Charlie, David, Emily", the program should print out "2".

str_list = list(map(str, input("Enter names separated by comma (', '): ").split(', ')))
count = 0

for i in str_list:
    if i.startswith('A') or i.startswith('E') or i.startswith('I') or i.startswith('O') or i.startswith('U'):
        count += 1

print(count)
```

```python
# 5. Write a function that takes a string as input and returns the string with all the vowels converted to uppercase. For example, if the input string is "hello world", the function should return "hEllO wOrld".

def vowel_to_upper(str1):
    new = ""
    for i in str1:
        if i in ("a", "e", "i", "o", "u"):
            new += i.upper()
        else:
            new += i
    return new

vowel_to_upper("hello world")
```

```python
# 26. Write a Python function that takes a string as input and returns a new string with all the words in reverse order. For example, if the input string is "hello world", the output should be "world hello".

def rev_str(str1):
    str_list = str1.split()
    str_list.reverse()
    return " ".join(str_list)

rev_str("hello world")
```
