```python
# 2. You have a list of grocery items and their prices. Write a Python code to calculate the total cost of the groceries.

groceries = [("Milk", 45), ("Bread", 30), ("Eggs", 60), ("Rice", 120), ("Apples", 80)]
total = 0
for item, price in groceries:
    total += price
print("Total cost of groceries:", total)
```

```python
# 3. Imagine you have a list of stock prices for a particular company. 
# Write a Python code to find the percentage increase or decrease in the stock price from the first day to the last day.

stock_prices = [150.0, 152.5, 148.0, 155.0, 160.0, 158.0, 162.0, 145.0]
change = (stock_prices[-1] - stock_prices[0]) * 100.00 / stock_prices[0]
print("Percentage:", change)
```

```python
# 4. You have a list of names and their ages. Write a Python code to find the name of the oldest person in the list.

people = [ {"name": "Alice", "age": 30}, {"name": "Bob", "age": 25}, 
          {"name": "Charlie", "age": 35}, {"name": "David", "age": 40}, 
          {"name": "Eve", "age": 35} ]
oldest_person_name = ""
oldest_age = 0
for p in people:
    if p["age"] > oldest_age:
        oldest_person_name = p["name"]
        oldest_age = p["age"]
print("Oldest person:", oldest_person_name)
```

```python
# 5. You have a list of strings representing sentences. 
# Write a Python code to count the number of words in each sentence and store them in a new list.

str_list = ["Life is so beautiful after the rain", "When you look back every youth is beautiful",
           "There is no end to your dreams journey but it's okay to take a break today",
           "Work hard on what you are given and you will see your path"]

freq = {}
for i in str_list:
    sen = i.split()
    for j in sen:
        if j not in freq:
            freq[j] = 1
        else:
            freq[j] += 1

count_list = list(freq.items())
print(count_list)
```

```python
# 6. Imagine you have a list of email addresses. 
# Write a Python code to create a new list that contains only the email addresses that end with ".com".

email_list = [
    "alice@example.com",
    "bob@domain.org",
    "carol@website.net",
    "dave@startup.com",
    "eve@company.co.in",
    "frank@service.com"
]
new_email = []

for i in email_list:
    if i[-4::1] == ".com":
        new_email.append(i)

print("Email addresses that end with '.com':", new_email)
```

```python
# 7. You have a list of numbers. Write a Python code to create a new list 
# that contains only the odd numbers from the original list.

number_list = [2, 5, 10, 13, 20, 25]
odd_num = []

for i in number_list:
    if i%2 != 0:
        odd_num.append(i)

print("Odd numbers list:", odd_num)
```

```python
# 13. Imagine you have a list of tuples where each tuple contains a name and a salary. 
# Write a Python code to create a new list that contains only the names of the people who earn more than $50,000 per year.

salary_data = [
    ("Anjali", 48000),
    ("Ravi", 52000),
    ("Sneha", 75000),
    ("Amit", 45000),
    ("Priya", 60000)
]
new_data = []

for i, j in salary_data:
    if j > 50000:
        new_data.append(i)

print("People who earn more than $50000 per year:", new_data)
```

```python
# 14. You have a list of strings representing sentences. Write a Python code to create a new list that 
# contains only the sentences that contain the word "Python" and "programming".

sentences = [
    "Python is a versatile language.",
    "I love Python programming.",
    "Programming in Java is fun.",
    "Python and programming go hand in hand.",
    "This course teaches Python basics."
]
new_sentence = []

for i in sentences:
    if "python" in i.lower() and "programming" in i.lower():
        new_sentence.append(i)

print("List that contains sentence having 'Python' and 'Programming':")
print(new_sentence)
```
