```python
# 1. Write a Python program that accepts a string and calculate the number of digits and letters.

str1 = input("Enter a string: ")
digit_count = 0
letter_count = 0

for i in str1:
    if i.isalpha():
        letter_count += 1
    if i.isdigit():
        digit_count += 1

print("Letters:", letter_count)
print("Digits:", digit_count)
```

```python
# 2. Write a Python program to check a triangle is equilateral, isosceles or scalene 

side1, side2, side3 = map(int, input("Enter three sides of triangle separated by space: ").split())

if side1 == side2 == side3:
    print("Equilateral Triangle")
elif side1 != side2 and side2 != side3 and side3 != side1:
    print("Scalene Triangle")
elif side1 == side2 or side2 == side3 or side3 == side1:
    print("Isosceles Triangle")
```

```python
# 4. Write a Python program to find the most common elements and their counts of a specified text. 

str1 = "lkseropewdssafsdfafkpwe"
freq = {}
freq_list = []

for i in str1:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

for i, j in freq.items():
    freq_list.append((i, j))

for i in range(len(freq_list)):
    for j in range(i+1, len(freq_list)):
        if freq_list[i][1] < freq_list[j][1]:
            freq_list[i], freq_list[j] = freq_list[j], freq_list[i]

print("Most common three characters of the said string:")
for i in range(3):
    print(freq_list[i])
```

```python
# 5. Write a Python program to compute the sum of first n given prime numbers. 

num = int(input("Enter a number (less than 10000) to compute the sum (0 to exit):"))
prime_num = []
i = 2

while len(prime_num) < num:
    is_prime = True
    if num == 0:
        break

    if num > 1:
        for j in range(2, i):
            if i%j == 0:
                is_prime = False
                break
    if is_prime == True:
        prime_num.append(i)
    i += 1

print(f"Sum of first {num} prime numbers: {sum(prime_num)}")
```

```python
# 6. Write a Python program to filter a dictionary based on values. 

stud_dict = {'Cierra Vega': 175, 'Alden Cantrell': 180, 'Kierra Gentry': 165, 'Pierre Cox': 190}
new_dict = {}

for i in stud_dict:
    if stud_dict[i] > 170:
        new_dict[i] = stud_dict[i]

print("Marks greater than 170:")
print(new_dict)
```
